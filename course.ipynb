```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

# ---------------------------------
# 1. Nhập đơn vị đo
# ---------------------------------
unit = input("Choose distance unit (km/miles): ").strip().lower()
u_label = "KM" if unit in ("km", "kilometer", "kilometers") else "MILES"

def get_val(val_km):
    return val_km if u_label == "KM" else val_km * 0.621371

# ---------------------------------
# 2. Xây dựng đồ thị bằng NetworkX (Graph Data Structure)
# ---------------------------------
G = nx.Graph()

# Tọa độ các trạm
pos = {
    "Yio Chu Kang": (4, 14), "Ang Mo Kio": (4, 11), 
    "Bishan": (4, 8), "Braddell": (4, 5), "Toa Payoh": (4, 2),
    "Marymount": (2, 6), 
    "Punggol": (14, 14), "Sengkang": (12, 12), 
    "Serangoon": (10, 10), "Woodleigh": (8, 8), "Potong Pasir": (6, 6),
    "Bedok North": (18, 10), "Macpherson": (16, 8), 
    "Geylang Bahru": (14, 6), "Bendemeer": (12, 4), "Jalan Besar": (10, 2),
    "Kembangan": (22, 10), "Eunos": (20, 8), 
    "Paya Lebar": (18, 6), "Aljunied": (16, 4), "Lavender": (14, 2)
}

# Thêm Nodes (Trạm) cùng thuộc tính pos (tọa độ) vào Graph
for node, coord in pos.items():
    G.add_node(node, pos=coord)

# Dữ liệu các tuyến đường (Edges) kèm khoảng cách theo km
edges_data = [
    # Red Line
    ("Yio Chu Kang", "Ang Mo Kio", 1.4), ("Ang Mo Kio", "Bishan", 1.6), ("Bishan", "Braddell", 1.2), ("Braddell", "Toa Payoh", 1.0),
    # Purple Line
    ("Punggol","Sengkang", 1.5), ("Sengkang","Serangoon", 5.1), ("Serangoon","Woodleigh", 1.2), ("Woodleigh","Potong Pasir", 0.9),
    # Blue Line
    ("Bedok North","Macpherson", 4.8), ("Macpherson","Geylang Bahru", 1.5), ("Geylang Bahru","Bendemeer", 1.7), ("Bendemeer","Jalan Besar", 0.9),
    # Green Line
    ("Kembangan","Eunos", 1.1), ("Eunos","Paya Lebar", 1.1), ("Paya Lebar","Aljunied", 1.3), ("Aljunied","Lavender", 2.9),
    # Orange Line (Circle Line)
    ("Marymount", "Bishan", 1.5), ("Bishan", "Serangoon", 2.5), ("Serangoon", "Macpherson", 3.5), ("Macpherson", "Paya Lebar", 1.5)
]

# Thêm Edges vào Graph kèm 2 thuộc tính 'km' và 'miles'
for u, v, w in edges_data:
    G.add_edge(u, v, km=w, miles=w * 0.621371)

interchanges = ["Bishan", "Serangoon", "Macpherson", "Paya Lebar"]

# ---------------------------------
# 3. Thông số Hình họa (Màu Sắc LTA)
# ---------------------------------
BG_COLOR = "#fbfcff"
MRT_COLORS = {"Red": "#EE2E30", "Green": "#009645", "Purple": "#9900AA", "Orange": "#FA9E0D", "Blue": "#005EC4"}
LW = 12 

fig, ax = plt.subplots(figsize=(20, 16), facecolor=BG_COLOR)
ax.set_facecolor(BG_COLOR)
ax.set_aspect('equal', adjustable='datalim') 

# ---------------------------------
# 4. Vẽ Các Tuyến (Extract từ tọa độ Node trong Graph)
# ---------------------------------
def draw_line(nodes, color):
    x = [G.nodes[n]['pos'][0] for n in nodes]
    y = [G.nodes[n]['pos'][1] for n in nodes]
    ax.plot(x, y, color=color, lw=LW, solid_capstyle='round', zorder=1)

draw_line(["Yio Chu Kang", "Ang Mo Kio", "Bishan", "Braddell", "Toa Payoh"], MRT_COLORS["Red"])
draw_line(["Punggol", "Sengkang", "Serangoon", "Woodleigh", "Potong Pasir"], MRT_COLORS["Purple"])
draw_line(["Bedok North", "Macpherson", "Geylang Bahru", "Bendemeer", "Jalan Besar"], MRT_COLORS["Blue"])
draw_line(["Kembangan", "Eunos", "Paya Lebar", "Aljunied", "Lavender"], MRT_COLORS["Green"])

# Khúc cua Circle Line (Orange)
t_arc = np.linspace(np.arccos(-0.8), np.arccos(0.8), 300)
x_arc = 10 + 10 * np.cos(t_arc)
y_arc = 10 * np.sin(t_arc)
ax.plot(x_arc, y_arc, color=MRT_COLORS["Orange"], lw=LW, solid_capstyle='round', zorder=1)

# ---------------------------------
# 5. Mác Trạm và Nhãn
# ---------------------------------
label_offsets = {
    "Yio Chu Kang": (-0.25, 0, 'right', 'center'), "Ang Mo Kio": (-0.25, 0, 'right', 'center'),
    "Braddell": (-0.25, 0, 'right', 'center'), "Toa Payoh": (-0.25, 0, 'right', 'center'),
    "Punggol": (-0.25, 0.25, 'right', 'bottom'), "Sengkang": (-0.25, 0.25, 'right', 'bottom'),
    "Woodleigh": (-0.25, 0.25, 'right', 'bottom'), "Potong Pasir": (0.3, -0.25, 'left', 'top'),
    "Bedok North": (-0.25, 0.25, 'right', 'bottom'), "Geylang Bahru": (-0.25, 0.25, 'right', 'bottom'),
    "Bendemeer": (-0.25, 0.25, 'right', 'bottom'), "Jalan Besar": (-0.25, 0.25, 'right', 'bottom'),
    "Kembangan": (-0.25, 0.25, 'right', 'bottom'), "Eunos": (-0.25, 0.25, 'right', 'bottom'),
    "Aljunied": (-0.25, 0.25, 'right', 'bottom'), "Lavender": (-0.25, 0.25, 'right', 'bottom'),
    "Bishan": (-0.45, 0.45, 'right', 'bottom'), "Serangoon": (-0.45, 0.45, 'right', 'bottom'),
    "Macpherson": (-0.5, 0, 'right', 'center'), "Paya Lebar": (0.45, -0.45, 'left', 'top'),
    "Marymount": (-0.25, -0.25, 'right', 'top')
}

# Mapping màu sắc cho các trạm tương ứng với màu line
node_colors = {
    "Yio Chu Kang": MRT_COLORS["Red"], "Ang Mo Kio": MRT_COLORS["Red"], 
    "Braddell": MRT_COLORS["Red"], "Toa Payoh": MRT_COLORS["Red"],
    "Punggol": MRT_COLORS["Purple"], "Sengkang": MRT_COLORS["Purple"], 
    "Woodleigh": MRT_COLORS["Purple"], "Potong Pasir": MRT_COLORS["Purple"],
    "Bedok North": MRT_COLORS["Blue"], "Geylang Bahru": MRT_COLORS["Blue"], 
    "Bendemeer": MRT_COLORS["Blue"], "Jalan Besar": MRT_COLORS["Blue"],
    "Kembangan": MRT_COLORS["Green"], "Eunos": MRT_COLORS["Green"], 
    "Aljunied": MRT_COLORS["Green"], "Lavender": MRT_COLORS["Green"],
    "Marymount": MRT_COLORS["Orange"]
}

# Xét duyệt vẽ Point trên bản đồ dựa trên Nodes của NetworkX Graph
for node in G.nodes():
    x, y = G.nodes[node]['pos']
    if node in interchanges:
        # Giữ nguyên node interchanges: màu trắng, viền đen
        ax.scatter(x, y, s=1200, color="white", edgecolors="black", linewidths=5, zorder=4)
        fs, fw = 15, "bold"
    else:
        # Sửa các node tuyến thành màu line tương ứng
        c = node_colors.get(node, "white")
        ax.scatter(x, y, s=400, color=c, edgecolors="#444", linewidths=3, zorder=3)
        fs, fw = 12, "bold"

    # In tên Ga tàu
    dx, dy, ha, va = label_offsets[node]
    ax.text(x+dx, y+dy, node.upper(), fontsize=fs, fontweight=fw, ha=ha, va=va, fontfamily='sans-serif',
            bbox=dict(facecolor=BG_COLOR, edgecolor='none', alpha=0.9, boxstyle='round,pad=0.1'))


# ---------------------------------
# 6. Gắn Nhãn Cự Ly (Trích Thuộc tính Cạnh Data từ Graph)
# ---------------------------------
def add_dist(x, y, val_km, val_miles):
    v = val_km if u_label == "KM" else val_miles
    ax.text(x, y, f" {v:.1f} {u_label} ", fontsize=11, fontweight="bold", color="white", ha='center', va='center',
            bbox=dict(facecolor="#4b5563", edgecolor="none", alpha=1.0, boxstyle="round,pad=0.3"), zorder=10)

angles = [np.arccos(-0.8), np.arccos(-0.6), np.pi/2, np.arccos(0.6), np.arccos(0.8)]
ccl_edge_angles = {
    tuple(sorted(("Marymount", "Bishan"))): 0,
    tuple(sorted(("Bishan", "Serangoon"))): 1,
    tuple(sorted(("Serangoon", "Macpherson"))): 2,
    tuple(sorted(("Macpherson", "Paya Lebar"))): 3
}

# Lấy cạnh (edges) và các thuộc tính từ Graph để lặp
for u, v, data in G.edges(data=True):
    edge_key = tuple(sorted((u, v)))
    
    # Xác định tọa độ hiển thị Textbox khoảng cách
    if edge_key in ccl_edge_angles:
        # Xử lí đặt độ dài theo cung cung tròn (CCL)
        i = ccl_edge_angles[edge_key]
        tm = (angles[i] + angles[i+1]) / 2
        lx = 10 + 10 * np.cos(tm)
        ly = 10 * np.sin(tm)
    else:
        # Đường thẳng mặc bản, khoảng cách nằm ở trung điểm 2 Ga
        ux, uy = G.nodes[u]['pos']
        vx, vy = G.nodes[v]['pos']
        lx, ly = (ux + vx) / 2, (uy + vy) / 2
        
    add_dist(lx, ly, data['km'], data['miles'])

# ---------------------------------
# 7. Legend Key 
# ---------------------------------
handles = [
    mlines.Line2D([0], [0], color=MRT_COLORS["Red"], lw=5, label="NSL (North-South Line)"),
    mlines.Line2D([0], [0], color=MRT_COLORS["Green"], lw=5, label="EWL (East-West Line)"),
    mlines.Line2D([0], [0], color=MRT_COLORS["Purple"], lw=5, label="NEL (North-East Line)"),
    mlines.Line2D([0], [0], color=MRT_COLORS["Orange"], lw=5, label="CCL (Circle Line)"),
    mlines.Line2D([0], [0], color=MRT_COLORS["Blue"], lw=5, label="DTL (Downtown Line)")
]
handles.append(mlines.Line2D([0], [0], marker='o', color='white', label='Interchange Station',
                             markeredgecolor='black', markeredgewidth=4, markersize=20, markerfacecolor='white'))

ax.legend(handles=handles, loc="upper right", frameon=True, fontsize=12, shadow=True, 
          title="MRT SYSTEM DIRECTORY", title_fontsize=14, labelspacing=1.3, borderpad=1.2, handlelength=1.5)

# ---------------------------------
# 8. Render Lên Hình & Save File (Task 1 Report)
# ---------------------------------
plt.title(f"MRT Singapore Network\n(All distances are displayed in {u_label})", 
          fontsize=26, fontweight="black", pad=35, color="#111")

plt.axis("off")
plt.tight_layout()
# Lưu ảnh ra để đưa vào bản PDF nộp
plt.savefig('task1_map.png', dpi=300, facecolor=BG_COLOR)
plt.show()

# ---------------------------------
# TASK 2: THỐNG KÊ (Tính Data từ NetworkX Graph)
# ---------------------------------
# Thao tác bắt buộc: Lấy dữ liệu thuộc tính thông qua Graph Edges Iterator
total_km = sum(data['km'] for u, v, data in G.edges(data=True))
total_miles = total_km * 0.621371
num_edges = G.number_of_edges()

avg_km = total_km / num_edges
avg_miles = total_miles / num_edges

print(f"\n{'='*55}\n      TASK 2: STATISTICAL CALCULATION\n{'='*55}")
print(f"1. Total Transportation Network Distance:")
print(f"   - {total_km:.1f} KM")
print(f"   - {total_miles:.2f} MILES")
print("-" * 55)
print(f"2. Average Distance between Stations:")
print(f"   - {avg_km:.2f} KM")
print(f"   - {avg_miles:.2f} MILES")
print("="*55)
```
